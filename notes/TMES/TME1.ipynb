```python
jo = "https://fr.wikipedia.org/wiki/Jeux_olympiques"

request_text = request.urlopen(jo).read()
print(request_text[:1000])
page = bs4.BeautifulSoup(request_text, "lxml")
print(page.find("title").text)
```

```python
mytable = page.find("table", class_ = "wikitable center")
body = mytable.find("tbody")
rows = body.find_all("tr")
#print(rows)
headers_with_th = rows[0].find_all("th")
headers=[cell.text.strip() for cell in headers_with_th]
#print(headers)
prv_year = "none"
prv_oly = "none"
data = []
for row in rows[1:]:
    line_with_td = row.find_all("td")
    line = [l.text.strip() for l in line_with_td]
    if len(line) == 5:
        prv_year = line[0]
        prv_oly = line[1]
        data.append(line)
    if len(line) == 3:
        if(line[0].isdigit()): # cas de lieu a determiner
            line +=[line[2],line[2]]
            data.append(line)
        else: #case of 2 rows collapsed
            liste1 = [prv_year,prv_oly ] + line
            data.append(liste1)
            
df = pandas.DataFrame(data, columns=headers)
display(df)
```

```python
import re

urls = []
indice = 0
page = bs4.BeautifulSoup(request.urlopen(jo).read(), "lxml")
set_cities = set([re.sub('\(\d\)',' ', row[2]).strip() for row in final_data[0:-1]])
for city in set_cities:
    a = page.find('a', attrs={"title": re.compile(city+'.*')})
    href = a['href']
    urls.append(href)
    
```

```python
from urllib.parse import urljoin

N = len(urls)
base_url = "https://fr.wikipedia.org"
while indice < N:
    url = urls[indice]
    absolute_url = urljoin(base_url, url)
    page = bs4.BeautifulSoup(request.urlopen(absolute_url).read(), "lxml")
    coordinates = page.find('span', id='coordinates')
    if coordinates:
        a = coordinates.find('a')
        if a:
            coord = a.text
            print(coord)
        else:
            print("No 'a' element found within coordinates.")
    else:
        print("No coordinates span found on the page.")
    indice += 1
    
    
```
